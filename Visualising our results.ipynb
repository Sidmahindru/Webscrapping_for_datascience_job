# Creating Visualization, Outlining Trends in Data Science required skills

Visualizations are created both for technical skills and management skills

```python
## Loading all the required libraries
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("results.csv")
df
```

```python
df = df.iloc[:,8:16]
df
```

```python
df = df.T
```

```python
df['No_of_ones'] = [11,0,1,3,0,7,9,0]
```

```python
df
```

```python
## grabing all the technical buzz words on the job description
gf = df[['No_of_ones']].plot(kind="bar", figsize=(20,10),title="Key Data Science Job Skills on Indeed")
gf.set_xlabel("Technical skills specified in Job Description")
gf.set_ylabel("No. of occurences")
plt.rcParams.update({'font.size':30})                                         
```
